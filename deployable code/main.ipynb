```python
def read_data():
      '''
        Reads different files
        Args: 
            df_prior_final :  Users prior order history
            df_train_final : Users last order data with product info
            df_train : Users last order data 
            df_orders : Order history 
            df_products_aisle_dep : DF of product,aisle,department info combined
        Returns:
                Featurized DF
    '''
    df_orders = pd.read_csv("../orders.csv")
    df_products = pd.read_csv("../products.csv")
    df_products_aisle_dep = pd.read_pickle("../df_products_aisle_dep.pkl")
    df_train = pd.read_csv("../order_products__train.csv")
    df_prior_final=  pd.read_csv('../df_prior_final.csv',dtype = {'order_id' : np.uint32,
                     'user_id' : np.uint32,
                    'product_id' : np.uint16,
                     'order_number' : np.uint8,
                     'reordered' : np.uint8,
                     'aisle_id' : np.uint8,
                     'department_id' : np.uint8,
                     'aisle' : 'category',
                     'department' : 'category',
                     'product_name' : 'category',
                     'add_to_cart_order' : np.uint8,
                     'eval_set' :'category',
                    'order_dow' : np.uint8,
                    'order_hour_of_day' : np.uint8
                })
    df_train_final = pd.read_csv('../df_train_final.csv',dtype = {'order_id' : np.uint32,
                     'user_id' : np.uint32,
                    'product_id' : np.uint16,
                     'order_number' : np.uint8,
                     'reordered' : np.uint8,
                     'aisle_id' : np.uint8,
                     'department_id' : np.uint8,
                     'aisle' : 'category',
                     'department' : 'category',
                     'product_name' : 'category',
                     'add_to_cart_order' : np.uint8,
                     'eval_set' :'category',
                    'order_dow' : np.uint8,
                    'order_hour_of_day' : np.uint8})
    return df_orders,df_products,df_products_aisle_dep,df_train,df_prior_final,df_train_final
```

```python
# reading related files
df_orders,df_products,df_products_aisle_dep,df_train,df_prior_final,df_train_final = read_data()
# Get feturized dataframe
data = feature_engineering.get_feature_engineering(df_prior_final,df_orders,df_products_aisle_dep)
# Prepare Train data
data_train = train_test_data.get_train_data(df_orders,df_train,data)
# Prepare test data
data_test = train_test_data.get_test_data(data)
#Train the model
model_train.train(data_train)
#Give predictions and generate submission.csv file
model_prediction.get_predictions(data_test,df_orders)
```
